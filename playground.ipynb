### Adjusted Rating

Works well only with ratings, not conversions.

```python
import pandas as pd
from plotly.graph_objects import *
import numpy as np
from scipy.stats import norm
```

```python
raw_rate = 7
global_mean = 5
n = 10

k = np.linspace(1, 1000, 1000)
ar = (raw_rate*n + global_mean*k)/(n + k)

fig = Figure(data = Scatter(x = k,
                            y = ar,
                            mode = 'markers'),
             layout = Layout(xaxis={'title': 'k'},
                             yaxis={'range': [4,7]},
                             title='Adjusted rating'))

fig.add_hline(y = global_mean)
fig.show()
```

```python
k = 10
global_mean = 5

r =  norm.rvs(7, 2, size = 1000)

adjusted_rating = [(np.mean(r[:i]*i + k*global_mean))/(i+k) for i in range(1, len(r))]
fig = Figure(data=Scatter(x=list(range(1,1000)),
                          y=adjusted_rating))

fig.show()
```

```python
k_vals = [10, 50]
global_mean = 5
traces = list()
r =  norm.rvs(7, 2, size = 1000)

for k in k_vals:
    adjusted_rating = [(np.mean(r[:i]*i + k*global_mean))/(i+k) for i in range(1, len(r))]
    traces.append(Scatter(x=list(range(1,1000)),
                          y=adjusted_rating,
                         name=str(k)))
fig = Figure(data=traces)
fig.show()
```
